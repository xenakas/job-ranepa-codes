```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
data_yearly = pd.read_csv('all_data_yearly.csv', index_col=None, parse_dates=True)
data_yearly_names = data_yearly.iloc[0,:]
data_yearly_scales = data_yearly.iloc[1,:]
data_yearly = data_yearly.iloc[2:,:]
data_yearly.index = data_yearly.index.astype(int)
data_yearly = data_yearly.astype(float)
```

```python
data_yearly['year']=data_yearly['year'].astype(int)
# data_yearly.index = data_yearly.index.astype(int)
```

```python
arrr = np.array([])
for i in np.arange(data_yearly.shape[0]):
    arrr = np.append(arrr, datetime.datetime(data_yearly.iloc[i,0], 6, 1, 0, 0) )
data_yearly.index = arrr

```

```python
data_fts = pd.read_csv('ФТС.csv', index_col=None, parse_dates=True)
data_fts_names = data_fts.iloc[0,:]
data_fts_scales = data_fts.iloc[1,:]
data_fts = data_fts.iloc[2:,:]
data_fts['year'] = data_fts['year'].astype(int)
data_fts['quarter'] = data_fts['quarter'].astype(int)
arrr = np.array([])
for i in np.arange(898):
    arrr = np.append(arrr, datetime.datetime(data_fts.iloc[i,1], data_fts.iloc[i,2], 1, 0, 0) )
data_fts.index = [data_fts['city'],arrr]
data_fts = data_fts.iloc[:,3:]
data_fts = data_fts.unstack(0)   
data_fts = data_fts.astype(float)
data_fts = data_fts.fillna(0)
```

```python
data_fts_sum = data_fts.sum(axis=1,level=0)
data_fi = data_fts.loc[:,'freight_in'].div(data_fts_sum['freight_in'], axis=0)
data_fo = data_fts.loc[:,'freight_out'].div(data_fts_sum['freight_out'], axis=0)
data_pi = data_fts.loc[:,'pass_in'].div(data_fts_sum['pass_in'], axis=0)
data_po = data_fts.loc[:,'pass_out'].div(data_fts_sum['pass_out'], axis=0)
data_ii = data_fts.loc[:,'ind_in'].div(data_fts_sum['ind_in'], axis=0)
data_io = data_fts.loc[:,'ind_out'].div(data_fts_sum['ind_out'], axis=0)
```

```python
data_ii.loc[:, (data_fi>0.01).sum() == 12]
data_io.loc[:, (data_fi>0.01).sum() == 12].iloc[:,2:5].sum(axis=1)
```

```python
dat1 = data_fts.loc[:,'freight_in'].loc[:, (data_fi>0.01).sum() == 12]
dat2 = data_fts.loc[:,'freight_out'].loc[:, (data_fi>0.01).sum() == 12]

fig, (ax1,ax2) = plt.subplots(2, figsize=(10, 8))

ax1.plot(dat1)
ax1.set_ylabel('Количество грузовых ВС')
ax1.set_title('В РОССИЮ')
ax1.legend(dat1.columns, loc='upper left', prop={'size': 10})
# ax1.set_xlabel('Год')
# ax1.tick_params(axis='y')
# ax1.set_ylim(3500, 3500)
ax2.plot(dat2)
ax2.legend(dat2.columns, loc='upper left', prop={'size': 10})
ax2.set_ylabel('Количество грузовых ВС')
ax2.set_title('ИЗ РОССИИ')
ax2.set_xlabel('Год')

# plt.show()
plt.savefig('all_plots/7_freight.png', dpi=1000)
```

```python
dat1 = data_fts.loc[:,'ind_in'].loc[:, (data_ii>0.003).sum() == 12]/1000
dat2 = data_fts.loc[:,'ind_out'].loc[:, (data_io>0.005).sum() == 12]/1000

fig, (ax1,ax2) = plt.subplots(2, figsize=(10, 8))

# ax.plot(data_yearly.loc[(data_yearly.index >= 1925)&(data_yearly.index <= 2000), ['n_freight_2','obor_gruz_2']])

color = 'tab:red'
# lns1 = 
ax1.plot(dat1)
ax1.set_ylabel('Тысяч человек')
ax1.set_title('В РОССИЮ')
ax1.legend(dat1.columns, loc='upper right', prop={'size': 10})
# ax1.set_xlabel('Год')
# ax1.tick_params(axis='y')
# ax1.set_ylim(3500, 3500)
ax2.plot(dat2)
ax2.set_ylabel('Тысяч человек')
ax2.set_title('ИЗ РОССИИ')
ax2.set_xlabel('Год')
ax2.legend(dat2.columns, loc='upper right', prop={'size': 10})


# plt.show()
plt.savefig('all_plots/7_people.png', dpi=1000)
```

```python
data_month = pd.read_csv('all_data_month.csv', index_col=None)
data_month_names = data_month.iloc[0,:]
data_month_scales = data_month.iloc[1,:]
data_month = data_month.iloc[2:,:]
data_month = data_month.astype(float)
data_month['year'] = data_month['year'].astype(int)
data_month['month'] = data_month['month'].astype(int)
arrr = np.array([])
for i in np.arange(291):
    arrr = np.append(arrr, datetime.datetime(data_month.iloc[i,0], data_month.iloc[i,1], 1, 0, 0) )
data_month.index = arrr
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

# ax.plot(data_yearly.loc[(data_yearly.index >= 1925)&(data_yearly.index <= 2000), ['n_freight_2','obor_gruz_2']])

color = 'tab:red'
lns1 = ax1.plot(data_month.loc[(data_month.index.year>2010)&(data_month.index.year<2020), ['ipc_gruz', 'ipc_pass']])
ax1.set_ylabel('процент к предыдущему месяцу')
ax1.set_xlabel('Год')
ax1.tick_params(axis='y')
# ax1.set_ylim(3500, 3500)

labs = [data_month_names['ipc_gruz'],data_month_names['ipc_pass']]
ax1.legend(labs, loc='upper left', prop={'size': 15})


# plt.show()
plt.savefig('all_plots/6_ipc.png', dpi=1000)
```
