## Прогнозы

### ~EIA~

В рамках прогноза EIA рассмтривается следующий  спектр потенциальных макроэкономических факторов, который может существенно повлиять на энергетические рынки в течение прогнозируемого периода [1]. Основным фактором, определяющим  неопределенность поставок энергоресурсов, является  влияние санкций на добычу нефти в России, решения ОПЕК+ по добыче и скорость, с которой производители нефти и природного газа в США будут наращивать  бурение.

Санкции в отношении России способствовали падению добычи нефти в России и продолжают создавать значительную неопределенность на рынке в отношении возможности дальнейших перебоев с поставками нефти.
Эти события произошли на фоне низких запасов нефти и постоянного повышения цен на нефть в связи с неопредленностью, связанной с распространением коронавирусной инфекции. 

Прогноз мирового потребления нефти и жидкого топлива в 2022 году был понижен   на 0,2 млн баррелей в сутки по сравнению с апрельским STEO, главным образом в результате понижения потребления роста в Китае и США. По прогнозам, мировое потребление нефти и жидкого топлива увеличится на 1,9 млн баррелей в сутки в 2023 году и составит в среднем 101,5 млн баррелей в сутки.

В рамках долгосрочного прогноза EIA [2]  рассмотрены следующие возможные  сценарии: (1) высокие цены на нефть, (2) низкие цены  на нефть, (3) высокое  предложение нефти и газа, (4) низкое предложение  нефти и газа, (5)  высокий экономический рост, (6)  низкий экономический рост, (7) высокие затраты  на ВИЭ, (8)  низкие затраты на ВИЭ.  Балансы мирового рынка, в первую очередь внешние факторы спроса и предложения, будут определять будущие цены на сырую нефть. В сценарии  высокой цены на нефть цена Brent может достигать  170 долларов за баррель, в базовом сценарии - 90 долларов за баррель, в случае низкой цены на нефть - 45  долларов за баррель .

Прогнозируемая добыча сырой нефти в США в Базовом сценарии достигает пика в конце 2020-х годов и остается близкой к этому уровню  до 2050 года.  По мере того, как мировая экономика будет возвращаться  к допандемическому уровню,  прогнозируется, что как спрос, так и цены останутся на высоком уровне, в результате чего добыча сырой нефти достигнет допандемического уровня в среднесрочной перспективе.




---

Ожидается, что средняя цена Brent будет во втором квартале 2022 составлять 107 дол. за баррель  и 103 дол. за баррель во второй половине 2022 года [1]. В 2023 году средняя цена упадет до 97 долларов за баррель. Однако в этом  ценовом  прогнозе существет большая доля неопределенности. 
В частности, EIA отмечает следующие ключевые факторы неопределенности:  

- воздействие санкций на Россию и продолжающийся эффект текущих санкций и действий частного сектора,
- потенциал новых санкций в отношении России, включая обсуждение общеевропейского запрета на импорт энергоносителей из России и темпы его реализации,
- рост спроса на нефть в течение лета при потенциальном  падении спроса из-за высоких розничных цен на топливо,
- объемы новой добычи сырой нефти, которая будет введена в эксплуатацию в рамках цен около 100 долларов США за баррель или выше,
- возможность возобновления всплесков случаев COVID-19 и характер ответных мер правительства,
- продолжающееся влияние скоординированного высвобождения запасов нефти из стратегических запасов в США и Европе,
- другие геополитические неопределенности, связанные с событиями в  Ливии,   Йемене или потенциальными новыми событиями в сделке с Ираном.


Фактические ценовые результаты будут во многом зависеть от того, в какой степени существующие санкции, введенные против России, повлияют на добычу нефти в России или продажу российской нефти на мировом рынке. 
Прогноз EIA был осущетсвлен 5 мая, поэтому он не учитывает запрет ЕС на импорт нефти из России. Тем не менее,  запреты введенные на момепнт написания прогноза  вероятно, будут способствовать более жесткому балансу нефти и росту  цена на нефть. Кроме того, степень реакции других производителей нефти на текущие цены на нефть и влияние макроэкономических изменений на мировой спрос на нефть будут иметь важное значение для формирования цен на нефть в ближайшие месяцы. 
Поскольку запасы нефти в настоящее время низки, ожидается, что давление в сторону снижения цен на нефть будет ограниченным и сложившиеся рыночные условия будут способствовать росту  волатильности цен.

В то же время потенциальное снижение спроса из-за таких факторов, как продолжающиеся жесткие меры по сдерживанию распространения COVID-19 в Китае, особенно в Шанхае, а также снижение опубликованной оценки ВВП США за первый квартал, способствовали снижению цен на сырую нефть по сравнению с мартом.

### ~OPEC~




Рост мирового спроса на нефть в 2021 году остается в целом неизменным 
на уровне 5,7 млн баррелей в сутки. Ожидается, что рост мирового спроса на нефть в 2022 году увеличится на 3,4 млн баррелей в сутки по сравнению с предыдущим годом. 
Восстановление спроса, поддерживается за счет  программам стимулирования и дальнейшего ослаблением мер по сдерживанию COVID-19 на фоне ускоренного развертывания вакцинации.


Прогнозируется, что рост спроса на нефть во втором квартале 2022 будет медленнее и составит 2,8 млн баррелей в сутки по сравнению с 5,2 млн баррелей в сутки в первом квартале 2022. 
Изменения отражают последние годовые данные, а также скорректированный в сторону повышения спрос на нефть в странах ОЭСР в результате улучшений экономической ситуации  в 4 квартале 2021 года. 

В 2022 году рост спроса на нефть был пересмотрен в сторону понижения на 0,3 млн баррелей в сутки до среднего уровня 3,4 млн баррелей в сутки в годовом исчислении с учетом потенциального снижения мирового ВВП и распространения омикрон-штамма COVID-19 в Китае и его влияния на мировой спрос на нефть. Ожидается, что на спрос в 2022 году повлияют текущие геополитические события в Восточной Европе, а также ограничения, связанные с пандемией COVID-19.


Ожидается, что дизельное топливо и бензин будут основными драйверами спроса на нефтепродукты в годовом исчислении, поскольку экономическая активность, мобильность и промышленная деятельность восстанавливаются во всем мире. Восстановление мобильности в сочетании со снижением ограничений, связанных с COVID-19, и устранением узких мест, связанных с торговлей, в основных странах-потребителях поддержит спрос на бензин и дизельное топливо, в то время как легкие дистилляты будут в значительной степени поддерживаться высоким спросом на нефтехимическую продукцию, особенно в Китае, США и Индии. Восстановление мировых авиаперевозок на фоне ослабления ограничений на поездки поддержит спрос на авиакеросин.

Спотовые цены на сырую нефть в апреле снизились. Замедление интереса к покупке сырой нефти на фоне ослабления опасений по поводу ужесточения рынка привело к резкой коррекции цен и росту  волатильности. 
Обеспеченность  рынка нефтью, наряду с наличием непроданных запасов, способствовали снижению спотовых цен. 


### BP

Роль нефти в  транспортном секторе падает во всех трех сценариях. В автомобильном транспорте это снижение в значительной степени отражает сочетание повышения эффективности транспортных средств и роста электрификации. В авиации и судостроении снижение добычи нефти обусловлено увеличением использования био- и водородного топлива.
Доминирующим фактором снижения потребления нефти автомобильным транспортом является повышение эффективности транспортных средств, особенно легковых автомобилей. 
К 2035 году 60–70 % сокращения потребления нефти в автомобильном транспорте по трем сценариям будет осущствлено за счет повышения эффективности транспортных средств и электрификации автомобильного парка. 


Использование нефти в авиации к 2050 году сократится примерно на 25% и 75%, соответственно, в сценариях "Accelerated" и "Net Zero", что вызыано  более широким  использованием экологичного авиационного топлива (SAF) в сочетании с повышением эффективности парка.  В "Accelerated" и "Net Zero" к 2050 году на нефть будет приходиться от 65% до 25% всей энергии, используемой в авиации, по сравнению с более чем 90% в "New Momentum".
В судостроении основной альтернативой нефти являются водородные топлива (аммиак, метанол и синтетическое дизельное топливо), на которые к 2050 году приходится от 30% до 55% всей конечной энергии в сценариях "Accelerated" и "Net Zero".

###  ~International Monetary Fund~

Цены на нефть и газ выросли на фоне войны на фоне собатий на Украине выросли на 36 процентов в период с августа 2021 года по февраль 2022 года, что было обусловлено сильным восстановлением спроса на нефть с краткосрочными эффектами варианта Omicron в конце 2021 года, за которым последовали геополитическая напряженность и начало спецоперации в феврале 2022 года. Сырая нефть марки Brent временно достигла 140 долларов в начале марта. Предложение нефти  было ограничено еще до событий на Украине, так как члены ОПЕК+  продолжали умеренными темпами сокращать ограничения поставок, а добыча в основных странах, не входящих в ОПЕК+, росла медленно. 

---

Ожидание снижения спроса на ископаемое топливо и, в меньшей степени, климатическая политика со стороны предложения (включая изменение общественных предпочтений в пользу устойчивого инвестирования) сократили капитальные затраты на нефть и газ во всем мире за последние три-четыре года. Этот факт  может оказывать постоянное повышательное давление на цены на нефть и другие виды ископаемого топлива и добавлять существенную неопределенность в прогнозы цен на нефть. 
Скоординированные климатические усилия стран-потребителей и производителей ископаемого топлива и отказ от ископаемого топлива темпами, соответствующими скорости внедрения возобновляемых источников энергии, помогут снизить риск высоких и неустойчивых цен на энергию. А меньшая политическая неопределенность поможет странам внести необходимые коррективы.



Отмечается, что рост субсидий в производство  электромобилей,  являются негативным шоком спроса на сырую нефть, поскольку сырая нефть заменяется электричеством, что приводит к снижению цен. Однако сокращение доли ископаемого топлива может также быть связано с ограничением притока инвестиций в нефть и газ из-за устойчивого инвестиционного давления и других мер политики в отношении предложения. 

Сокращение добычи нефти, которое гипотетически вызвано только мерами со стороны предложения, вместо этого будет оказывать сильное повышательное давление на цены, приводя к росту цен примерно до 190 долларов за баррель, что принесет пользу странам-производителям за счет потребителей.  Поскольку добыча нефти была бы прибыльной для всех производителей, основными факторами, определяющими распределение производства и ренты, были бы ограничения страны, экологические нормы и доступ к капиталу.

США к 2025 году, по причине замедления роста мировой экономики и постепенной адаптации мирового энергетического рынка к новым условиям, в частности замещения российских энергоносителей на рынках "недружественных" стран, с одной стороны, и перенаправление российского энергетического экспорта в "нейтральные" другие страны - с другой. 
В консервативном варианте мировые цены на нефть «Юралс»
падают до уровней около 57 долл. США за баррель к 2025 году.

### ~Банки~


По прогнозам  Wells Fargo добыча нефти, скорее всего, останется невосприимчивой к текущим высоким ценам. Не смотря на ожидаемую высокую волатильность цен на нефть, увеличение предложения производителями в ответ на скачок цен не прогнозируется. 
В результате наши целевые диапазоны цен на сырую нефть на конец года были повышены, чтобы учесть увеличивающийся дисбаланс между спросом и предложением. Спрос на нефть уже вырос и ожидается, что он продолжит расти, если ограничения, связанные с COVID-19, ослабнут и спрос на нефть на рынке авиации будет восстанавливаться.




Аналитики Morgan Stanley повысили прогноз цены на нефть Brent в третьем квартале 2022 года до 130 дол./бар.
Предполгается, что спрос на нефть в 2022 году будет ниже, чем прогнозировалось ранее.
Тем не менее, его влияние на цены будет более чем компенсировано эмбарго на российскую нефть, самостоятельными ограничениями на ее закупку сырьевыми трейдерами и более низким ростом поставок из Ирана.
Более высокие цены на сырьевые товары, конфликт на Украине, Covid-19 в Китае и перспектива ужесточения денежно-кредитной политики Федеральной резервной системы будут ограничивать спрос.

По прогнозам ABN Amro ожидается рост цен на нефть, особенно во второй половине года. 
После снятия карантина в Китае, спрос на нефть в Китае восстановится и усилит давление на и без того напряженные рынки. 

Bank of America (BoA) прогнозирует, что цены на нефть марки Brent могут подняться выше 150 долларов за баррель, если произойдет резкое сокращение экспорта российской нефти.
Аналитики BoA не прогнозируют восстановления спроса на нефть до уровня, существовавшего до пандемии COVID-19, в 2022 году, поскольку сохраняются проблемы с поставками.
В 2023 году спрос на нефть может приблизиться к докризисному уровню, если добыча жидких углеводородов  в России будет поддерживаться на уровне около 10 миллионов баррелей в сутки, а поставки ОПЕК+ увеличатся.



Barclays также повысил прогноз цен на нефть, заявив, что недавно объявленные данные по стратегическим нефтяным запасам не могут быть устойчивым источником поставок и не окажут длительного влияния на цены. 
Barclays также снизил свой прогноз роста спроса на 2022 год на 550 000 баррелей в сутки (баррелей в сутки), сославшись на последствия замедления роста экономики, ограничений на мобильность, связанных с COVID в Китае и более высоких затрат на продукты питания и энергию для потребительских расходов.


JP Morgan понизил свой прогноз мирового спроса на нефть на 2022 год на один миллион баррелей в сутки, сославшись на более высокие цены на нефть, ухудшение перспектив роста и эскалацию геополитической напряженности. Прогноз цены на нефть марки Brent сохранился на уровне 114 долларов за баррель во втором квартале этого года.
Отмечается, что цены на нефть застряли в узком диапазоне, поскольку слабый мировой спрос на нефть и высвобождение стратегических запасов нивелируют растущие ожидания того, что Европа может согласиться ограничить закупки нефти у России, говорится в сообщении банка.
Способность России перенаправить свои нефтяные потоки добавляет неорпделенности в прогнозы.


Из-за шока предложения, что ожидается, что цены на нефть  должны вырасти до 120 долларов за баррель и оставаться на этом уровне в течение нескольких месяцев, чтобы стимулировать падение спроса, при условии отсутствия дополнительных  объемов  нефти из Ирана. 
Оценивается, что из-за напряженности на нефтяном рынке потеря еще одного миллиона баррелей в сутки может привести к росту цен на нефть на 18-35 долларов за баррель выше целевой цены.
Иранская сделка же может немедленно увеличить поставки на 1 млн баррелей в сутки в течение следующих двух месяцев за счет высвобождения плавучих хранилищ.
