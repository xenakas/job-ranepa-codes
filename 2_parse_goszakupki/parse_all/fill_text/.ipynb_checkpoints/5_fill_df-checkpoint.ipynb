```python
def read_table(table):
    data = []
    keys = None
    for i, row in enumerate(table.rows):
        text = (cell.text for cell in row.cells)
        if i == 0:
            keys = tuple(text)
            continue
        row_data = dict(zip(keys, text))
        data.append(row_data)
    return pd.DataFrame(data) 
```

```python
def read_table_ext(table):
    t_l = len(table.rows)
    t_w = max([len(table.rows[i].cells) for i in range(len(table.rows))])
    t_emp = np.array(['']*t_l*t_w)
    data = pd.DataFrame(t_emp.reshape(-1, t_w))
    for i, row in enumerate(table.rows):
        k = 1
        for j, cell in enumerate(row.cells):
            data.iloc[i,j] = cell.text    
            if (j!=0) & (data.iloc[i,j-k] == cell.text):
                data.iloc[i,j] = ''
                k += 1    
    return data
```

```python
def read_table_ext_2(table):
    t_l = len(table.rows)
    t_w = max([len(table.rows[i].cells) for i in range(len(table.rows))])
    t_emp = np.array(['']*t_l*t_w)
    data = pd.DataFrame(t_emp.reshape(-1, t_w))
    for i, row in enumerate(table.rows):
        k = 1
        k2 = 1
        for j, cell in enumerate(row.cells):

            data.iloc[i,j] = cell.text    

            if (j!=0) & (data.iloc[i,j-k] == cell.text):
                data.iloc[i,j] = ''
                k += 1

            if len(cell.tables) != 0:
                for i in range(len(cell.tables)):
                    data_2 = read_table_ext(cell.tables[i])
#                     print(k2)
                    if (k2==1):
                        data = data.append(data_2, ignore_index=True)
                        k2 +=1
                    if (k2!=1):
                        if not(data_2.reset_index(drop=True).equals(data.iloc[-data_2.shape[0]:, :(data_2.shape[1])].reset_index(drop=True))):
                            data = data.append(data_2, ignore_index=True)

            
    return data.fillna('')
```
